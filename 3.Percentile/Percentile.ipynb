```python
#call function

def quanQual(dataset):
    qual=[]
    quan=[]
    for columnName in dataset.columns:
        if(dataset[columnName].dtype=='O'):
            #print("Qual")
            qual.append(columnName)     #append is to add into the list of columnname
        else:
           # print("Quan")
            quan.append(columnName)
    return quan,qual
```

#Percentile

```python
np.percentile(dataset["ssc_p"],50)   #percentile is the value exist within the range
```

```python
descriptive=pd.DataFrame(index=["mean","median","mode","Q1:25%","Q2:50%","Q3:75%","Q4:100%"],columns=quan)
for columnName in quan:
    descriptive.loc["mean",columnName]=dataset[columnName].mean()
    descriptive.loc["median",columnName]=dataset[columnName].median()
    descriptive.loc["mode",columnName]=dataset[columnName].mode()[0]
```

```python
descriptive
```

```python
descriptive=pd.DataFrame(index=["mean","median","mode","Q1:25%","Q2:50%","Q3:75%","Q:99%","Q4:100%"],columns=quan)
for columnName in quan:
    descriptive.loc["mean",columnName]=dataset[columnName].mean()
    descriptive.loc["median",columnName]=dataset[columnName].median()
    descriptive.loc["mode",columnName]=dataset[columnName].mode()[0]
    descriptive.loc["Q1:25%",columnName]=dataset.describe()[columnName]["25%"]
    descriptive.loc["Q2:50%",columnName]=dataset.describe()[columnName]["50%"]
    descriptive.loc["Q3:75%",columnName]=dataset.describe()[columnName]["75%"] #Nan(salary) value should be preprocessed so here took from describe() function
    descriptive.loc["Q:99%",columnName]=np.percentile(dataset[columnName],99)  #method to call percentile function
    descriptive.loc["Q4:100%",columnName]=dataset.describe()[columnName]["max"]
```
