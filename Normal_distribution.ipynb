```python
def get_pdf_probabilities(dataset,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    ax=sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')   #kde is the kernel density if it True the curve will reflect
    pyplot.axvline(startrange,color='Red')   #axis vertical line of the start range
    pyplot.axvline(endrange,color='Red')
    sample=dataset
    sample_mean=sample.mean()
    sample_std=sample.std()
    print('mean={:.3f},standard Deviation={:.3f}'.format(sample_mean,sample_std))
    dist=norm(sample_mean,sample_std)
    values=[value for value in range(startrange,endrange)]
    probabilities=[dist.pdf(value) for value in values]
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(20,30,sum(probabilities)))
    return prob
```

```python
get_pdf_probabilities(dataset["ssc_p"],50,75)   #if start range and endrange gives same value the probability is "zero" because there is no density/
```

```python
# standard Normal distribution

import seaborn as sns
mean=dataset["ssc_p"].mean()
std=dataset["ssc_p"].std()
values=[i for i in dataset["ssc_p"]]
```

```python
mean
```

```python
std
```

```python
z_score=[((j-mean)/std) for j in values]  # z_score is used to convert the standardized value by (x-mean/std)
```

```python
z_score
```

```python
import numpy as np  # converting the z_score list method into array method by using numpy to find the average value for z_score
z=np.array(z_score)  
z.mean()       #always in Standard normal distibution the mean is "0" and std is "1" here the mean value is -4.62 is equal to 0 (i,e. common in numpy and pandas)
```

```python
sum(z_score)/len(z_score)   #without converting get the average of z_score.
```

```python
round(z.mean(),-5)  # by using round function we get the exact value
```

```python
def stdNBgraph(dataset):
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()
    values=[i for i in dataset]
    z_score=[((j-mean)/std) for j in values]
    sns=sns.distplot(z_score,kde=True)
    sum(z_score)/len(z_score)
```

```python
stdNBgraph(dataset["ssc_p"])
```
